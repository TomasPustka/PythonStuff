```python
import pandas
from geopy.geocoders import Nominatim
```

```python
df = pandas.read_json("supermarkets.json")
```

```python
df["Address"] = df["Address"] + ", " + df["City"] + ", " + df["State"] + ", " + df["Country"]
```

```python
df
```

```python
nom = Nominatim(user_agent = "http")
```

```python
df["Coordinates"] = df["Address"].apply(nom.geocode)
```
